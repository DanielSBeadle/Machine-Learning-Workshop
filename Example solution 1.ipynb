**Task 1**

```python
import pandas 

df = pandas.read_csv("dataset.csv")
df
```

**Task 2**

```python
df['type'] = df['type'].replace({'white': 1, 'red': 0})
df
```

**Task 3**

```python
from sklearn.model_selection import train_test_split

y = df.loc[:,"quality"]
x = df.drop(["quality"], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

**Task 4**

```python
from sklearn.dummy import DummyRegressor

dummy = DummyRegressor(strategy='mean')
dummy.fit(x_train, y_train)
```

**Task 5**

```python
training_predictions = dummy.predict(x_train)
test_predictions = dummy.predict(x_test)
```

**Task 6**

```python
from sklearn.metrics import mean_squared_error

training_error = mean_squared_error(y_train, training_predictions)
print('Dummy regressor training error:' ,training_error)
test_error = mean_squared_error(y_test, test_predictions)
print('Dummy regressor test error:' ,test_error)
```

**Task 7**

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train, y_train)

training_predictions = model.predict(x_train)
test_predictions = model.predict(x_test)

training_error = mean_squared_error(y_train, training_predictions)
print('Training error:', training_error)

test_error = mean_squared_error(y_test, test_predictions)
print('Test error:', test_error)
```
